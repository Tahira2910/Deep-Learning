```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Convert 'diagnosis' column to categorical data type
data['diagnosis'] = pd.Categorical(data['diagnosis'])
```

```python
# X--> ind, Y--> Dep
X = data.iloc[:, 2:].values
Y = data.iloc[:, 1].values

# Encoding categorical data
from sklearn.preprocessing import LabelEncoder
labelencoder_X_1 = LabelEncoder()
Y = labelencoder_X_1.fit_transform(Y)
```

```python
# Splitting the dataset into training and test set
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y , test_size = 0.1, random_state = 42)
```

```python
# Feature Scaling

# StandardScaler--> to standardize the data, to normalize the data
# values btw +1 and -1
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
import keras
# Sequential--> classification technique based on step by step(In sequence)
from keras.models import Sequential
# Dense--> use for adding layers, input layer, hidden layer...
from keras.layers import Dense
```

```python
# adding the input and first hidden layer
classifier = Sequential()
# ind variable quantity: so,input_dim=30, hidden layer--> output layer, units: 16-->(31/2)
# 30 ind variable quan, one predicted: 30+1 = 31/2
# initialization: weight value-->uniform distribution
classifier.add(Dense(units = 16, input_dim=30, kernel_initializer='uniform', activation = 'relu'))

# adding the second hidden layer
classifier.add(Dense(units=16, kernel_initializer='uniform', activation = 'relu'))

# adding the output layer, units
# prediction will be on 1 output layer, unit
# activation funtion on output layer, unit is sigmoid
classifier.add(Dense(units=1, kernel_initializer='uniform', activation = 'sigmoid'))
```

```python
# Now, we will compile our classifier
# output prediction is in binary(B or M) so, using binary_crossentropy
# and for multiple, categorical_crossentropy
classifier.compile(optimizer='Adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
classifier.fit(X_train,Y_train,batch_size=100, epochs=150)
# classifier.fit(X_train,Y_train,batch_size=8, epochs=5)
```

```python
X_test
```

```python
# Predicting the Test set results
Y_Pred = classifier.predict(X_test)
Y_Pred = (Y_Pred > 0.5)
```

```python
# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(Y_test, Y_Pred)
```
